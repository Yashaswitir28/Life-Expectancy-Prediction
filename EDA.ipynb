```python
df.columns = df.columns.str.strip()
def find_outliers(df, column):
    if df[column].dtype not in [np.float64, np.int64]:
        print(f"Column '{column}' is not numeric. Skipping outlier detection.")
        return 0, 0

    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    num_outliers = len(outliers)
    percent_outliers = (num_outliers / len(df)) * 100 if len(df) > 0 else 0

    return num_outliers, percent_outliers

# Ensure column names are clean
df.columns = df.columns.str.strip()

# List of columns to check
columns_to_check = [
    'life expectancy', 'Adult Mortality', 'infant deaths', 'Alcohol',
    'percentage expenditure', 'Hepatitis B', 'Measles', 'under-five deaths',
    'Polio', 'Total expenditure', 'Diphtheria', 'HIV/AIDS', 'GDP',
    'Population', 'thinness  1-19 years', 'thinness 5-9 years',
    'Income composition of resources', 'Schooling'
]

# Iterate through columns and handle missing columns gracefully
for column in columns_to_check:
    if column in df.columns:
        num_outliers, percent_outliers = find_outliers(df, column)
        print(f"---------------{column}---------------")
        print(f"Number of outliers: {num_outliers}")
        print(f"Percent of data that is outlier: {percent_outliers:.2f}%")
    else:
        print(f"Column '{column}' not found in the dataset. Skipping.")
```
